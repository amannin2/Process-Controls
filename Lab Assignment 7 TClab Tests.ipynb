```python
%matplotlib inline
from tclab import TCLab, clock, Historian, Plotter
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import least_squares
```

### Part 2: Test Optimization Estimates using TCLab

```python
with TCLab() as lab:
    
    t_final = 1000
    t_step = 5

    # set up historian and plotter
    h = Historian(lab.sources)
    p = Plotter(h, t_final)

    # initialize manipulated variable
    lab.P1 = 200
    lab.P2 = 100

    # event loop
    for t in clock(t_final, t_step):
            lab.Q1(38.53)
            lab.Q2(38.5)
            p.update(t)     
```

#### Discussion

With the given heater settings from the optimization problem, T1 = 46 degrees and T2 = 42 degrees compared to the expected values of T1 = 45 degrees and T2 = 40 degrees. The optimization model overestimates the required heater temperature for both heaters, but the model is within 3 degrees of the expected temperature values.  

### Part 3: Test Max Temperature Difference

```python
with TCLab() as lab:
    
    t_final = 800
    t_step = 5

    # set up historian and plotter
    h = Historian(lab.sources)
    p = Plotter(h, t_final)

    # initialize manipulated variable
    lab.P1 = 200
    lab.P2 = 100

    # event loop
    for t in clock(t_final, t_step):
            lab.Q1(77.73)
            lab.Q2(0)
            p.update(t)
```

#### Discussion

With the given heater settings from the optimization problem, T1 = 60 degrees and T2 = 41 degrees compared to the expected values of T1 = 60 degrees and T2 = 39.84 degrees. In this case, the heater values developed from the optimization system work extremely well in generating the maximum temperature difference between the two heaters. The maximum temperature difference occurs where the heater with the higher power input is set to the maximum possible temperature of the system (60 degrees C) and the other heater is kept off for the entire run.
